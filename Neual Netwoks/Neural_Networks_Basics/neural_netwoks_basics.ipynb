✔ 1. Neurons (Nodes)

These are like individual processing units.
Each neuron receives inputs, calculates something, and passes output.

```python
✔ 2. Layers

Input → Hidden → Output
```

✔ 3. Weights

Every connection between neurons has a weight.
Weight decides the strength of a connection.
Output = Inputs × Weights

```python
✔ 4. Bias

Bias shifts the output.
Similar to intercept (c) in equation y = mx + c.
Purpose: improves model flexibility.
```

✔ 5. Activation Function

Introduces non-linearity.

Examples:
1.ReLU
2.Sigmoid
3.Tanh
4.Softmax
*Without activation → neural network behaves like linear equation (useless).

✔ 6. Forward Propagation

How data flows through layers:
Input → Hidden → Output
Neural network calculates predictions.

 ✔ 7. Loss Function

Measures how wrong the prediction was.
For example:
Squared error loss (Regression)
Cross entropy loss (Classification)
Loss tells:
➡ “How bad is my model?”

✔ 8. Backpropagation

Neural network calculates gradient of loss w.r.t. weights.
It understands how much each weight contributed to error.

✔ 9. Gradient Descent

Optimization algorithm:
Update weights to reduce loss

✔ 10. Learning Rate

Controls how big the weight update step is.
Too high → unstable
Too low → slow

VISWAL SUMMARY

Inputs ---> [Weights + Bias] ---> Activation ---> Prediction
              ↑         |
              |         V
       Backpropagation (Corrections)
